```python
year_data =[]
player_data =[] 
position_data = []
team_data = []

all_data = soup.find("table", class_="tablehead")
rows = all_data.find_all("tr")[2:]
for row in rows:
    col = row.find_all("td")
    year = col[0].text.strip()
    player = col[1].text.strip()
    position = col[2].text.strip()
    team = col[3].text.strip()
    year_data.append(year)
    player_data.append(player)
    position_data.append(position)
    team_data.append(team)
    
```

```python
year_data
```

```python
data = {
        'Year' :year_data,
        'Player Name' :player_data,
        'Position' :position_data,
        'Club' :team_data,
}
pd.DataFrame(data).to_excel('NbaMvpAwardAllTime.xlsx')
```
